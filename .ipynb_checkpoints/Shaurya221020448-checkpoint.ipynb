```python
df["Land Area(Km2)"] = df["Land Area(Km2)"].str.replace(',','').astype(float)
```

```python
mean = df["Land Area(Km2)"].mean()
```

```python
df["Land Area(Km2)"].fillna(mean, inplace = True)
```

```python
df.head()
```

```python
def changeandmean(df, lst, ch):
    df[lst] = df[lst].str.replace(ch,'').astype(float)
    mean = df[lst].mean()
    df[lst].fillna(mean, inplace = True)
```

```python
changeandmean(df, "Co2-Emissions", ',')
changeandmean(df, "Armed Forces size", ',')
changeandmean(df, "Agricultural Land( %)", '%')
changeandmean(df, "Population: Labor force participation (%)", '%')
changeandmean(df, "Tax revenue (%)", '%')
changeandmean(df, "Unemployment rate", '%')
changeandmean(df, "CPI", ',')
changeandmean(df, "CPI Change (%)", '%')
changeandmean(df, "Gasoline Price", '$')
changeandmean(df, "Minimum wage", '$')
changeandmean(df, "Gross primary education enrollment (%)", '%')
changeandmean(df, "Gross tertiary education enrollment (%)", '%')
```

```python
df.dropna(inplace = True)
```

```python
df.to_csv("Cleaned Country Dataset.csv", index = False)
```

# OneHot Encoding

```python
pd.concat([dm['Country'], temp], axis = 1).head()
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
dm['Capital/Major City'].unique()
```

```python
dm['le_Capital/Major City'] = le.fit_transform(dm['Capital/Major City'])
```

```python
dm.head()
```

# Frequency Encoding

```python
value_counts = dm['Capital/Major City'].value_counts().to_dict()
value_counts
```

```python
dm['Capital/Major City'] = dm['Capital/Major City'].map(value_counts)
dm.head()
```

```python
frequency_count = (dm['Capital/Major City'].value_counts() / len(dm)).to_dict()
frequency_count
```

```python
dm['Capital/Major City'] = dm['Capital/Major City'].map(frequency_count)
dm.head()
```

# Mean Encoding

```python
dm.groupby(['Abbreviation'])['Land Area(Km2)'].mean()
```

```python
mean_labels = dm.groupby(['Abbreviation'])['Land Area(Km2)'].mean().to_dict()
mean_labels
```

```python
dm['Abbreviation_mean'] = dm['Abbreviation'].map(mean_labels)
dm
```
