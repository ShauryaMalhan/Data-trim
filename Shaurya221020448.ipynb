```python
df["Land Area(Km2)"] = df["Land Area(Km2)"].str.replace(',','').astype(float)
```

```python
mean = df["Land Area(Km2)"].mean()
```

```python
def changeandmean(df, lst, ch):
    df[lst] = df[lst].str.replace(ch,'').astype(float)
    mean = df[lst].mean()
    df[lst].fillna(mean, inplace = True)
```

```python
changeandmean(df, "Co2-Emissions", ',')
changeandmean(df, "Armed Forces size", ',')
changeandmean(df, "Agricultural Land( %)", '%')
changeandmean(df, "Population: Labor force participation (%)", '%')
changeandmean(df, "Tax revenue (%)", '%')
changeandmean(df, "Unemployment rate", '%')
changeandmean(df, "CPI", ',')
changeandmean(df, "CPI Change (%)", '%')
changeandmean(df, "Gasoline Price", '$')
changeandmean(df, "Minimum wage", '$')
changeandmean(df, "Gross primary education enrollment (%)", '%')
changeandmean(df, "Gross tertiary education enrollment (%)", '%')
```

# Outlier Trimming

```python
plt.rcParams["figure.figsize"] = [8,6]
sns.set_style("darkgrid")
```

```python
sns.boxplot( y='Armed Forces size', data=dm)
```

```python
IQR = dm["Armed Forces size"].quantile(0.75) - dm["Armed Forces size"].quantile(0.25)
lower_age_limit = dm["Armed Forces size"].quantile(0.25) - (IQR * 1.5)
upper_age_limit = dm["Armed Forces size"].quantile(0.75) + (IQR * 1.5)
print(lower_age_limit)
print(upper_age_limit)
```

```python
age_outliers = np.where(dm["Armed Forces size"] > upper_age_limit, True, np.where(dm["Armed Forces size"] < lower_age_limit, True, False))
```

```python
age_outliers[1:40]
```

```python
Armed_Forces_size_outliers = dm.loc[~(age_outliers), ]
dm.shape, Armed_Forces_size_outliers.shape
```

```python
Armed_Forces_size_outliers[30:40]
```

```python
sns.boxplot( y='Armed Forces size', data = Armed_Forces_size_outliers)
```
